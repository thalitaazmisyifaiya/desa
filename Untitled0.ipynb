```python
import re, random, numpy as np, pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, f1_score, precision_recall_fscore_support,
                             confusion_matrix, ConfusionMatrixDisplay)
import torch
from datasets import Dataset, DatasetDict
from transformers import (T5ForConditionalGeneration, T5Tokenizer,
                          Trainer, TrainingArguments)
```

```python
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

print("CUDA available?", torch.cuda.is_available())
```

```python
df["content"] = (df["judul"].astype(str).fillna("") + ". " +
                 df["narasi"].astype(str).fillna(""))
```

```python
def clean_text(s: str) -> str:
    s = re.sub(r"http\S+|www\.\S+", " ", s)   # Hapus URL
    s = re.sub(r"<[^>]+>", " ", s)            # Hapus tag HTML
    s = re.sub(r"[^a-zA-Z\s]", " ", s)        # Hanya huruf & spasi
    s = s.lower()                             # Ubah ke huruf kecil
    s = re.sub(r"\s+", " ", s).strip()        # Hapus spasi berlebih
    return s
```

```python
df["content_clean"] = df["content"].apply(clean_text)
```

```python
label_map = {0: "TRUE", 1: "FAKE"}  # sesuaikan jika label dataset berbeda
df["target_text"] = df["label"].map(label_map)
```

```python
df["input_text"] = "detect: " + df["content_clean"]
```

```python
df_final = df[["input_text", "target_text"]].copy()

print("Contoh data hasil pra-pemrosesan:")
display(df_final.head())
```

```python
train_df, temp_df = train_test_split(
    df_final, test_size=0.20, stratify=df_final["target_text"], random_state=SEED
)
val_df, test_df  = train_test_split(
    temp_df, test_size=0.50, stratify=temp_df["target_text"], random_state=SEED
)

print(f"Train: {len(train_df)}, Val: {len(val_df)}, Test: {len(test_df)}")
```

```python
train_ds = Dataset.from_pandas(train_df.reset_index(drop=True))
val_ds   = Dataset.from_pandas(val_df.reset_index(drop=True))
test_ds  = Dataset.from_pandas(test_df.reset_index(drop=True))
dataset  = DatasetDict({"train": train_ds, "validation": val_ds, "test": test_ds})
```

```python
MODEL_NAME = "t5-small"
tokenizer = T5Tokenizer.from_pretrained(MODEL_NAME)
```

```python
def tokenize_function(batch):
    model_inputs = tokenizer(
        batch["input_text"], max_length=512, padding="max_length", truncation=True
    )
    labels = tokenizer(
        batch["target_text"], max_length=4, padding="max_length", truncation=True
    )
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

tokenized_ds = dataset.map(tokenize_function, batched=True,
                           remove_columns=["input_text", "target_text"])
tokenized_ds.set_format("torch")
```

```python
# tokenized_ds = dataset.map(tokenize_function, batched=True, remove_columns=["input_text", "target_text"])
```

```python
model = T5ForConditionalGeneration.from_pretrained(MODEL_NAME)

def normalize_label(s: str) -> str:
    s = re.sub(r"\s+", "", s).upper()
    if s in ["TRUE"]: return "TRUE"
    if s in ["FAKE"]: return "FAKE"
    return "FAKE"

id2cls = {"TRUE": 1, "FAKE": 0}
```

```python
def compute_metrics(eval_pred):
    preds, labels = eval_pred
    decoded_preds  = tokenizer.batch_decode(preds, skip_special_tokens=True)
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)
    y_pred = [id2cls.get(normalize_label(p), 0) for p in decoded_preds]
    y_true = [id2cls.get(normalize_label(l), 0) for l in decoded_labels]
    acc = accuracy_score(y_true, y_pred)
    f1w = f1_score(y_true, y_pred, average="weighted")
    prec, rec, f1_each, _ = precision_recall_fscore_support(
        y_true, y_pred, labels=[1,0], zero_division=0
    )
    return {
        "accuracy": acc,
        "f1_weighted": f1w,
        "precision_TRUE": float(prec[0]),
        "recall_TRUE": float(rec[0]),
        "f1_TRUE": float(f1_each[0]),
        "precision_FAKE": float(prec[1]),
        "recall_FAKE": float(rec[1]),
        "f1_FAKE": float(f1_each[1]),
    }
```

```python
from transformers import AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer
training_args = Seq2SeqTrainingArguments(
    output_dir="/content/t5_results",
    eval_strategy="epoch",
    save_strategy="epoch",
    predict_with_generate=True,
    generation_max_length=4,
    learning_rate=5e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_steps=50,
    load_best_model_at_end=True,
    metric_for_best_model="f1_weighted",
    greater_is_better=True,
    seed=SEED
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_ds["train"],
    eval_dataset=tokenized_ds["validation"],
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
trainer.save_model("/content/t5_news_detector")
tokenizer.save_pretrained("/content/t5_news_detector")
```

```python
eval_test = trainer.evaluate(eval_dataset=tokenized_ds["test"], metric_key_prefix="test")
print("Hasil evaluasi di Test Set:", eval_test)
```

```python
pred_test = trainer.predict(tokenized_ds["test"])
decoded_preds  = tokenizer.batch_decode(pred_test.predictions, skip_special_tokens=True)
labels = np.where(pred_test.label_ids != -100, pred_test.label_ids, tokenizer.pad_token_id)
decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)
y_pred = [id2cls.get(normalize_label(p), 0) for p in decoded_preds]
y_true = [id2cls.get(normalize_label(l), 0) for l in decoded_labels]

cm = confusion_matrix(y_true, y_pred, labels=[1,0])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["TRUE","FAKE"])
disp.plot(values_format="d")
plt.title("Confusion Matrix – Testing Set")
plt.show()
```

```python
def predict_news(text: str) -> str:
    prompt = "classify the news as TRUE or FAKE: " + clean_text(text)
    input_ids = tokenizer.encode(prompt, return_tensors="pt", max_length=512, truncation=True)
    if torch.cuda.is_available():
        model.to("cuda")
        input_ids = input_ids.to("cuda")
    outputs = model.generate(input_ids, max_length=4)
    pred = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return "TRUE" if normalize_label(pred) == "TRUE" else "FAKE"
```

```python
ontoh_true = "Presiden mengumumkan kebijakan baru terkait pajak mulai Januari 2025."
contoh_fake = "Ilmuwan membuktikan bumi datar setelah penelitian 10 tahun."
print("Prediksi contoh TRUE:", predict_news(contoh_true))
print("Prediksi contoh FAKE:", predict_news(contoh_fake))
```
